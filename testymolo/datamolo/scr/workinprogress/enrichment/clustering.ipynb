```python
import subprocess
import os
```

```python
with open('in/sequence.fasta', 'w') as handle:
    with open('data/db/Nidovirales.fasta') as nido:
        with open('data/vaz2.json/tmp.fasta') as tmp:
            handle.write(tmp.read() + '\n' + nido.read())

# create db
command:str = "mmseqs createdb in/sequence.fasta in/seq"     
# cluster db
command:str = "mkdir out/cluster"
command:str = "mmseqs cluster in/seq in/seq out/cluster/clu tmp"
# clu_seq -> fasta
command:str = "mkdir out/cluster/seq"
command:str = "mmseqs createseqfiledb in/seq out/cluster/clu out/cluster/seq/clu.seq"

command:str = "mkdir out/cluster/seq/fasta" 
command:str = "mmseqs result2flat in/seq in/seq out/cluster/seq/clu.seq out/cluster/seq/fasta/clu.fasta"      
```

```python
# is header 
def isHeader(line:str) -> bool :
    if line.startswith('>'):
        return True
    return False

clu:list = []
wasHeader:bool = False  ## if previous line was also header
fasta:str = ""

with open ("out/clu.fasta") as handle :
    for line in handle.readlines():

        if isHeader(line):
            if wasHeader:
                fasta = ""

            else:
                if len(clu) > 0: 
                    clu[-1].append(line.strip() + '\n' + fasta)
                
            wasHeader = True
            if not fasta == "":
                clu.append([]) 
            fasta = line.strip() + '\n'

        else:
            fasta = line.strip() + '\n'
            wasHeader = False
    #end
    clu[-1].append(fasta)
                
```

```python
print("nombre of clusters", len(clu))
for group in clu:
    print(len(group), group)
```
